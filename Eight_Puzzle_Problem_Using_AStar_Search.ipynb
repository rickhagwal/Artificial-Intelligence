```python
#implementation of class State, class EightPuzzle and function "print_result(result)"
import math
class State():
   # Build Successors to previous state by sliding tile from given index
    def __init__(self , tiles , index_zero , g , h , prev , slide_from_index,type_val,astar_type):
        eight_puzzle=EightPuzzle()
            
        if type_val=='initial':
            self.tiles = tiles
            self.index_zero = index_zero
            self.g = g
            self.h = h
            self.prev = prev
            self.slide_from_index = slide_from_index
        else:
            self.tiles = prev.tiles.copy()
            self.tiles[prev.index_zero]=self.tiles[slide_from_index]
            self.tiles[slide_from_index] = 0
            self.index_zero = slide_from_index
            self.g = prev.g + 1

            if astar_type=='misplaced':
                self.h = eight_puzzle.misplacedTilesHeuristic(self.tiles,eight_puzzle)
            else:
                self.h = eight_puzzle.manhattanDistanceHeuristic(self.tiles,eight_puzzle)
            self.prev = prev
            
    def __hash__(self):
        return hash((self.tiles, self.index_zero, self.g , self.h, self.prev, self.slide_from_index))

    def __eq__(self, other):
        if not isinstance(other, type(self)): return NotImplemented
        return self.tiles == other.tiles and self.index_zero == other.index_zero and self.g == other.g  and self.h == other.h  and self.prev == other.prev 
    
    def __lt__(self, other):
        if self.f(self.g,self.h) == other.f(other.g,other.h):
            return other.g-self.g
        else:
            return self.f(self.g,self.h) - other.f(other.g,other.h)
        
    # Goal Test to test whether it is Goal or not
    def f(self,g,h):
        return g + h
        
    def move_up( self,tiles,index_zero, g , h , prev,astar  ):
        if index_zero > 2:
            return State (tiles , index_zero , g , h , prev , index_zero - 3 ,'moveup' ,astar)
        else:
            return None
        
    def move_down( self,tiles,index_zero, g , h , prev ,astar):
        if index_zero < 6:
            return State (tiles , index_zero , g , h , prev , index_zero + 3 ,'movedown',astar )
        else:
            return None
            
    def move_left( self,tiles,index_zero, g , h , prev ,astar):
        if index_zero % 3 > 0:
            return State(tiles , index_zero , g , h , prev ,  index_zero - 1 ,'moveleft',astar )
        else:
            return None
        
    def move_right( self,tiles,index_zero, g , h , prev,astar ):
        if index_zero % 3 < 2:
            return State(tiles , index_zero , g , h , prev ,  index_zero + 1 ,'moveright',astar )
        else:
            return None
        
    def print_state( self,tiles ):
        print("f = " , f() , " = g+h = " , g , "+" , h)
        for i in range(0,8,3):
            print(tiles[i] , " " , tiles[i+1] , " " , tiles[i+2])
        
    def print_all( self,tiles ):
        if (prev != null): prev.printAll()
        print( '\n' )
        print_state(tiles)
        
class EightPuzzle:   
    
    def addSuccessor( self, successor,explored_set,frontier_queue ):
        if successor != None and successor.tiles not in explored_set:
            frontier_queue.put(successor)
        
    def misplacedTilesHeuristic(self,tiles,eight_puzzle):
        h=0
        for i in range(len(tiles)):
            if tiles[i]!=0:
                h = h + eight_puzzle.misplacedTiles(i, tiles[i])
        return h
        
    def misplacedTiles(self,a,b):
        if a!=b:
            return 1
        return 0
    
    def manhattanDistanceHeuristic(self,tiles,eight_puzzle):
        h=0
        for i in range(len(tiles)):
            if tiles[i]!=0:
                h = h + eight_puzzle.manhattanDistance(i, tiles[i])
        return h
        
    def manhattanDistance(self,a,b):
        return math.fabs(a / 3 - b / 3) + math.fabs(a % 3 - b % 3);
    
    def index(self,tiles,val):
        for i in range(len(tiles)):
            if tiles[i] == val: return i
        return -1
    
    def solve(self , tiles , explored_set , frontier_queue ,initial_state , eight_puzzle , goal_state , index_zero, g ,h,astar): 
        
        while not frontier_queue.empty():
            try:
                frontier_queue.get(False)
            except Empty:
                continue
            frontier_queue.task_done()
            
        explored_set.clear()
        
        # Add initial state to queue.
        frontier_queue.put(initial_state,0)
        
        while not frontier_queue.empty():

#             // Get the lowest priority state.
            state_low_prior = frontier_queue.get()
#             // If it's the goal, we're done.
            if (state_low_prior.tiles == goal_state): 
                print("Average_Steps = ",len(explored_set))
                state_low_prior.print_all(tiles)
                return
            
  #           // Make sure we don't revisit this state.
            if state_low_prior.tiles not in explored_set:
                explored_set.append(state_low_prior.tiles)
            print ('state_low_prior.tiles ',state_low_prior.tiles)
            index_zero=eight_puzzle.index(state_low_prior.tiles,0)
   #          // Add successors to the queue.
            eight_puzzle.addSuccessor(state_low_prior.move_up(tiles, index_zero, g , h, state_low_prior,astar),explored_set,frontier_queue);
            eight_puzzle.addSuccessor(state_low_prior.move_down(tiles, index_zero, g , h, state_low_prior,astar),explored_set,frontier_queue);
            eight_puzzle.addSuccessor(state_low_prior.move_right(tiles, index_zero, g , h, state_low_prior,astar),explored_set,frontier_queue);
            eight_puzzle.addSuccessor(state_low_prior.move_left(tiles, index_zero, g , h, state_low_prior,astar),explored_set,frontier_queue);

```
