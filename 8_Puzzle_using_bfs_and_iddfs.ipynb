```python
#implementation of function "print_result(result)"
def print_result(result):
    for i in range(len(result)-1):
        state = list(map(int, result[i]))
        print_states(state)
        print('\nto\n')
    state = list(map(int, result[len(result)-1]))
    print_states(state)
    
def print_states(state):
    print (state[0], state[1], state[2])
    print (state[3], state[4], state[5])
    print (state[6], state[7], state[8])

#implementation of class Node"
#Through the node class, both IDS and BFS can be called
import sys
import time
class Node:
    def __init__(self,data=None,next_node=None):
        self.data = data
        self.next_node = next_node
    
    def get_data(self):
        return self.data
    
    def get_next(self):
        return self.next_node
    
    def set_next(self,new_node):
        self.next_node = new_node
    
    #Moves
    def move_up( self,current_state,explored_set,state_history,state,frontier,new_value,search_algo,start_time):
        index = current_state.index( '0' )
        if index > 2:
            child_state = current_state[0:index-3]+"0"+current_state[index-2:index]+current_state[index-3]+current_state[index+1:]
            state.goal_test(child_state,current_state,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
        
    def move_down( self,current_state,explored_set,state_history,state,frontier,new_value,search_algo,start_time):
        index = current_state.index( '0' )
        if index < 6:
            child_state = current_state[0:index]+current_state[index+3:index+4]+current_state[index+1:index+3]+"0"+current_state[index+4:]
            state.goal_test(child_state,current_state,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
        
    def move_left( self,current_state,explored_set,state_history,state,frontier,new_value ,search_algo,start_time):
        index = current_state.index( '0' )
        if index != 0 and index != 3 and index!= 6:
            child_state = current_state[0:index-1]+"0"+current_state[index-1]+current_state[index+1:]
            state.goal_test(child_state,current_state,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
    
    def move_right( self,current_state,explored_set,state_history,state,frontier,new_value ,search_algo,start_time):
        index = current_state.index( '0' )
        if index != 2 and index != 5 and index != 8:
            child_state = current_state[0:index]+current_state[index+1]+"0"+current_state[index+2:]
            state.goal_test(child_state,current_state,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
    
    #Add child state to Frontier, Explored Set and State History
    def add(self,child_state,parent_state,explored_set,state_history,frontier,new_value,search_algo):
        if child_state not in explored_set:
            if search_algo == 'idsSearch':      # Iterative Deepening DFS
                frontier.insert(0, child_state) #insert element at starting of list-FILO(Stack), using list
            else:                               # BFS
                frontier.add(child_state)       #insert element at starting of list-FIFO(Queue), using set
            if parent_state is None:
                new_value = 0
            else:
                new_value = explored_set.get(parent_state)+ 1
            state_history[child_state] = parent_state
            explored_set[child_state] = new_value
    
    #Goal Test to check whether goal state is reached yet or not
    def goal_test(self,child_state,parent_state,explored_set,state_history,state,frontier,new_value,search_algo,start_time):
        state.add(child_state,parent_state,explored_set,state_history,frontier,new_value,search_algo)
        goal='012345678'
        if child_state==goal:
            if search_algo == 'idsSearch':
                search_algo='Iterative Deepening DFS'
            else:
                search_algo='BFS'
            print('Total Length of explored_set', len( explored_set))
            trace_node = child_state
            list_nodes= []
            print('Average_Steps taken by ',search_algo,' : ' ,explored_set[trace_node])
            while trace_node is not None:
                list_nodes.append(trace_node)
                trace_node = state_history[trace_node]
            list_nodes.reverse()
            print_result(list_nodes)
            print ('Average_Time taken by ',search_algo,' : ' ,(time.time() - start_time))
            sys.exit(0)        
            
#implementation of function "Iterative_deepening_DFS"    
def main(search_algo):
    start_time = time.time()
    input_node='108537642'
    depth_limit=30 #Considered, depth limit as 30
    solution=False
    state = Node()
    frontier = []
    state_history = {}
    explored_set = {}
    new_value=0
    parent_state=None
    state.add(input_node,parent_state,explored_set,state_history,frontier,new_value,search_algo)
    while frontier is not None: #
        current_node=frontier.pop() #deletes element from the end of frontier-FILO(Stack), using list
        if explored_set.get(current_node) < depth_limit:
            state.move_up(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
            state.move_down(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
            state.move_left(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
            state.move_right(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)


#implementation of function "breadthFirstSearch" (for graduate students)
def main(search_algo):
    start_time = time.time()
    input_node='108537642'
    state = Node()
    frontier = set()
    state_history = {}
    explored_set = {}
    new_value=0
    parent_state=None
    state.add(input_node,parent_state,explored_set,state_history,frontier,new_value,search_algo)
    while frontier is not None:
        current_node=frontier.pop() #deletes element from the starting of frontier-FIFO(Queue), using set
        state.move_up(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
        state.move_down(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
        state.move_left(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
        state.move_right(current_node,explored_set,state_history,state,frontier,new_value,search_algo,start_time)
    print('Solution doesnt exist')

if __name__ == "__main__":
    main('bfs')
#main('idsSearch')
```
